```python
setosa = iris[iris["Species"]=="Iris-setosa"]
```

```python
virginica = iris[iris["Species"]=="Iris-virginica"]
```

```python
iris['Species'].unique()
```

```python
versicolor = iris[iris["Species"]=="Iris-versicolor"]
```

```python
pt = iris['PetalLengthCm']
```

# UNIVARIATE ANALYSIS


```python
plt.plot(versicolor['PetalLengthCm'],np.zeros_like(versicolor['PetalLengthCm']),'bo')
plt.plot(virginica['PetalLengthCm'],np.zeros_like(virginica['PetalLengthCm']),'go')
plt.plot(setosa['PetalLengthCm'],np.zeros_like(setosa['PetalLengthCm']),'ro')
```
