```python
data_set = pd.read_csv("day.csv", index_col = "instant" )
```

```python
data_set
```

```python
X = data_set[['season','mnth','holiday','weekday','workingday','weathersit','temp','atemp','hum','windspeed']].values
```

```python
data_set['rental'] = data_set["casual"]
```

```python
data_set
```

```python
data_set = data_set.drop('casual', axis=1)
```

```python
data_set
```

```python
y = data_set['rental'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
len(X_test),len(X_train),len(X)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
pred = lr.predict(X_test)
```

```python
pred
```

```python
y_test
```

```python
from sklearn.metrics import mean_squared_error , r2_score
```

```python
mean_squared_error(y_test,pred)
```

```python
r2_score(y_test,pred)
```

```python
from sklearn.linear_model import Lasso

lasso = Lasso().fit(X_train,y_train)

y_pred = lasso.predict(X_test)

r2_score(y_test,y_pred)
```

```python
from sklearn.tree import DecisionTreeRegressor

dt = DecisionTreeRegressor().fit(X_train,y_train)

y_pred_dt = dt.predict(X_test)

r2_score(y_test,y_pred_dt)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor().fit(X_train,y_train)

y_pred_rf = rf.predict(X_test)

r2_score(y_test,y_pred_rf)
```

```python
from sklearn.ensemble import GradientBoostingRegressor

gb = GradientBoostingRegressor().fit(X_train,y_train)

y_pred_gb = gb.predict(X_test)

r2_score(y_test,y_pred_gb)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer
alg = GradientBoostingRegressor()
params = {
    'learning_rate' : [0.1,0.1111111,0.2,0.12],
    'n_estimators' : [100,150,170,200,400]
}
score = make_scorer(r2_score)

gridsearch = GridSearchCV(alg,params,scoring=score,cv=5,return_train_score=True)
gridsearch.fit(X_train,y_train)

gridsearch.best_params_ , gridsearch.best_score_
```

```python
from sklearn.ensemble import GradientBoostingRegressor

gb = GradientBoostingRegressor(learning_rate = 0.12, n_estimators = 100).fit(X_train,y_train)

y_pred_gb = gb.predict(X_test)

r2_score(y_test,y_pred_gb)
```
