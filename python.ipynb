```python
def prime_num(num):
    if num>1:
        if num == 2:
            return True
        for i in range(2,num):
            if num%i==0:
                return False
        return True
    elif num == 1:
        return False
```

```python
import numpy as np
lst = np.arange(0,20)
```

```python
lst
```

```python
# MAPPING FUNCTION
list(map(prime_num,lst))
```

```python
def primenumbers(num):
    if num>1 :
        if num == 2:
            return f"{num} is a prime number "
    for i in range (2,num):
        if num%i==0:
            return f"{num } is not a prime number "
    else :
        return f"{num } is a prime number "
```
